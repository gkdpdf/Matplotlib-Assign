#2
Scatter plot is used to show the correlation between the two variables

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Title")
```

#4
With the subplot() function you can draw multiple plots in one figure.

```python
import numpy as np


x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2= np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])


x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])



x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, ax = plt.subplots(2, 2)
ax[0][0].plot(x1,y1)
ax[0][1].plot(x2,y2)
ax[1][0].plot(x3,y3)
ax[1][1].plot(x4,y4)
```

#5
Bar plots are a type of data visualization used to represent data in the form of rectangular bars.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

#5
a box plot or boxplot is a method for graphically demonstrating the locality, spread and skewness groups of numerical data through their quartiles

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
fig,ax=plt.subplots(2)

ax[0].boxplot(box1)
ax[1].boxplot(box2)
```
